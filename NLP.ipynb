```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


a=pd.read_csv(r"/Users/vijay/Downloads/nlp-getting-started/train.csv")
a
```

```python
a=a[['text','target']]
a
```

```python
data=np.array(a)[:,0]
data
```

```python
y=np.array(a)[:,1]
y=y.astype('int')
y
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
vect = CountVectorizer()
vect.fit(data)
X = vect.transform(data) 

feature_names = vect.get_feature_names()
print("Number of features: {}".format(len(feature_names)))
print("First 100 features:\n{}".format(feature_names[:100])) 
print("Features 20080 to 20090:\n{}".format(feature_names[20080:20090]))
```

```python
X.shape
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
scores = cross_val_score(LogisticRegression(max_iter=10000), X, y, cv=5) 
print("Mean cross-validation accuracy: {:.2f}".format(np.mean(scores)))
```

```python
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10,100]}
grid = GridSearchCV(LogisticRegression(max_iter=10000), param_grid, cv=5) 
grid.fit(X,y)
print("Best cross-validation score: {:.2f}".format(grid.best_score_))
print("Best parameters: ", grid.best_params_)
```

```python
# Let's remove noise first, by removing all elements that appear less than 5 times

vect = CountVectorizer(min_df=5).fit(data) 
X1 = vect.transform(data)
X1

feature_names = vect.get_feature_names()
print("Number of features: {}".format(len(feature_names)))
print("First 100 features:\n{}".format(feature_names[:100]))
print("Features 2000 to 2020:\n{}".format(feature_names[2000:2020]))
```

```python
# after df=5 , let's see the results
scores = cross_val_score(LogisticRegression(max_iter=10000), X1, y, cv=5) 
print("Mean cross-validation accuracy: {:.2f}".format(np.mean(scores)))
```

```python
# How about grid search for parameter C
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10,100]}
grid = GridSearchCV(LogisticRegression(max_iter=10000), param_grid, cv=5) 
grid.fit(X1,y)
print("Best cross-validation score: {:.2f}".format(grid.best_score_))
print("Best parameters: ", grid.best_params_)
```

```python
# So cleaning noise hasn't improved accuracy
# how about tfid ?
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(TfidfVectorizer(min_df=5, norm=None),LogisticRegression(max_iter=10000))
param_grid = {'logisticregression__C': [0.001, 0.01, 0.1, 1, 10]}
grid = GridSearchCV(pipe, param_grid, cv=5)
grid.fit(data,y)
print("Best cross-validation score: {:.2f}".format(grid.best_score_))
```

```python
# no difference after tfid
# how about n-gram?

```

```python
vect = CountVectorizer(min_df=5,ngram_range=(1, 3)).fit(data) 
X1 = vect.transform(data)
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10,100]}
grid = GridSearchCV(LogisticRegression(max_iter=10000), param_grid, cv=5) 
grid.fit(X1,y)
print("Best cross-validation score: {:.2f}".format(grid.best_score_))
print("Best parameters: ", grid.best_params_)

```

```python
# so no change, after n-gram maximum accuracy is 0.73
# how about english-stopwords?
vect = CountVectorizer(min_df=5, stop_words="english").fit(data)
X1 = vect.transform(data)
grid = GridSearchCV(LogisticRegression(max_iter=10000), param_grid, cv=5) 
grid.fit(X1,y)
print("Best cross-validation score: {:.2f}".format(grid.best_score_))
print("Best parameters: ", grid.best_params_)
```

```python
# accuracy decreases after using stopwords
```

```python
# So every method to improve accuracy fails here 
```
